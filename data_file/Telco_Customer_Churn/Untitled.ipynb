```python
import os
import json

# 假设 absolute_path 已定义
description_path =  "parsed_description.json"

# 1. 读取 JSON 文件
with open(description_path, "r", encoding="utf-8") as f:
    data = json.load(f)  # data 是字典，包含 root 键


data.pop("Churn")  # 安全删除

# 3. 写回修改后的数据
with open(description_path, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)  # 保持缩进格式
```
